```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_excel("data.xlsx")

print ((data["Meas act SPECT"]-data["Real act"])/data["Real act"])*100

act_cons = np.unique(data["Act conc"])

fig1 = plt.figure()
ax = fig1.add_subplot(111)

for i in range(len(act_cons)):

    act_1 = data[data["Act conc"] == act_cons[i]]

    label_str = str(act_cons[i])
    label_str = label_str[0:4] + " MBq/ml"
    
    ax.plot(act_1["Real act"], act_1["Meas act SPECT"], 'o', label = label_str)

plt.legend(loc = 2)
ax.plot(act_1["Real act"], act_1["Real act"], '--')

    
plt.show()
```
